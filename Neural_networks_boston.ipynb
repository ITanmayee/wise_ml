```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/Boston.csv")
data.head()
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df, train_df['medv'], test_df['medv'] = train_test_split(data.drop(['medv'], axis=1), data['medv'], test_size=0.2, random_state=50, shuffle=True)
train_df.head()
```

```python
predictors = train_df.columns[:-1]  # excluding the target column 'medv'
target = 'medv'
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))   # scaling all the data within the range 0 to 1 to avaid variation in the values
scaled_train = scaler.fit_transform(train_df)
scaled_train  # fit_transform returns a array

```

```python
scaled_train_df = pd.DataFrame(scaled_train, columns=train_df.columns.values)  # converting the array into dataframes and the column names are named as per the original data
scaled_train_df
```

Model building


```python
# Sequential Layer
model = Sequential()
```

```python
# defining 3 hidden layers

model.add(Dense(units=50, activation='relu'))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=50, activation='relu'))
```

```python
# Output Layer
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
X = scaled_train_df.drop(target, axis=1).values
y = scaled_train_df[[target]].values
```

```python
# Train the model

model.fit(
    x=X,
    y=y,
    epochs=100,
    shuffle=True,
    verbose=2,
)

# batch_size can be used when you have a larger data set
```

```python
def scaler_inverse_transform(value):
    scaled_prediction = value
    scaled_prediction -= added_factor
    return scaled_prediction / multiply_factor
```

```python
multiply_factor = scaler.scale_[13]
added_factor = scaler.min_[13]
```

```python
scaler_inverse_transform(y[:1][0][0])
```

```python
prediction = model.predict(X[:])
prediction
```

```python
for idx in range(404):    
    print(f"Actual Price: {scaler_inverse_transform(y[idx][0])* 1000:.2f}$")
    print(f"Predicted Price: {scaler_inverse_transform(prediction[idx][0])* 1000:.2f}$")
    print("-"*30)
```
